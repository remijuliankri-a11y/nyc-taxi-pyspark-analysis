### Q1. Latency vs Throughput
Latency is the time it takes for data to move through a system - from the moment something is submitted until the result is produced. It tells us how quickly a single task is completed. Throughput, on the other hand, describes how much data a system can handle within a given period of time, so it reflects overall capacity or efficiency.

In short, latency is about the speed of each task, while throughput is about the volume processed over time.

A practical example is a streaming data pipeline. Lower latency allows near-real-time analytics, while higher throughput is essential when large amounts of data (such as IoT sensor streams) need to be ingested continuously without causing bottlenecks. In many big-data systems like Hadoop or Spark, a small trade-off in latency is acceptable if the goal is to handle massive datasets efficiently.

### Q2. Role of HDFS in Big Data Architecture
HDFS (Hadoop Distributed File System) is the storage foundation of the Hadoop ecosystem, designed to store and manage massive datasets across multiple low-cost servers. It provides a distributed, fault-tolerant, and high-throughput storage layer, enabling scalable and reliable data processing for frameworks such as MapReduce and Spark.

In a Big Data architecture, HDFS ensures that large volumes of data are split into blocks, replicated across several DataNodes for reliability, and centrally managed by the NameNode, which maintains metadata and block locations. This design allows organizations to store petabytes of structured and unstructured data efficiently, using affordable hardware while ensuring data availability and recovery in case of node failure.

By acting as the distributed storage layer beneath processing frameworks like YARN and Spark, HDFS enables both batch and real-time analytics within the broader Hadoop architecture.

### Q3. ETL vs ELT (+ examples)
ETL (Extract, Transform, Load) and ELT (Extract, Load, Transform) are two data integration approaches that differ mainly in where and when data transformations occur.

ETL performs transformations before loading the data into the destination system. I would say it is best suited for traditional data warehouses where data must be cleaned, standardized, and structured in a staging area prior to loading. For example, a company might use ETL to prepare transactional data from multiple OLTP systems before loading it into a relational warehouse for business intelligence dashboards.

ELT, by contrast, loads raw data first and then performs transformations inside the destination system, typically within a data lake or cloud platform that supports large-scale parallel processing. ELT is ideal for Big Data and unstructured data, as it allows faster ingestion and greater scalability - for instance, when loading IoT or log data directly into a cloud data lake for on-demand transformation and machine learning.

In essence, ETL prioritizes data quality and structure before storage, while ELT prioritizes speed, scalability, and flexibility, making it the modern standard for cloud-based analytics.

### Q4. RDBMS vs Hadoop vs Spark
Traditional RDBMS systems process structured data using SQL and ACID-compliant transactions, making them ideal for operational workloads such as banking, ERP, and CRM systems. They scale primarily through vertical scaling (stronger hardware), which limits their ability to handle the volume, velocity, and variety characteristic of Big Data.

Hadoop, by contrast, was designed for large-scale distributed data processing. Using HDFS for storage and MapReduce for computation, Hadoop can handle structured, semi-structured, and unstructured data across thousands of commodity nodes. It relies on horizontal scaling and built-in fault tolerance, making it suitable for batch analytics, large data archiving, and data warehousing workloads that require high throughput over low latency.

Apache Spark represents a more modern, in-memory processing engine that supports both batch and streaming analytics with significantly lower latency than MapReduce. Sparks in-memory Resilient Distributed Datasets (RDDs) and high-level APIs (Spark SQL, MLlib, Structured Streaming) allow fast iterative processing and ML at scale. It also scales horizontally and excels in real-time use cases like IoT analytics, fraud detection, and interactive ETL pipelines.

Overall, RDBMS is best for transactional integrity and smaller structured datasets, Hadoop excels at large-scale batch processing and storage, and Spark delivers high-speed, flexible analytics for both real-time and large distributed workloads.

### Q5. Structured vs Semi-Structured vs Unstructured
Structured data follows a strict schema, is stored in predefined formats like tables with rows and columns and is labeled. It is easier to query with SQL and common in OLTP and data warehouse systems (e.g. customer records, transactions, sensor readings). Because of its rigid organization, structured data offers high accuracy, consistency, and fast retrieval.

Semi-structured data does not use a fixed relational schema but still has tags, hierarchies, or metadata that provide partial structure (some labeled but mostly unlabeled). I would say that JSON, XML, log files, API responses, and emails are typical examples. Semi-structured data is more flexible than structured data and widely used in modern web applications, NoSQL databases, and ETL/ELT pipelines.

Unstructured data has no predefined format or schema and is unlabeled. This includes free text, images, video, audio, social media posts, and scanned documents. It requires advanced techniques (like NLP, computer vision, or distributed systems such as Spark) to extract insights. It makes up most Big Data and is often used in analytics, machine learning, and digital transformation.

To summarize it: structured data is rigid and easy to query, semi-structured data is flexible with partial organization, and unstructured data is raw and schema-free, requiring more advanced processing.

### 1.1 Automated Download Script for NYC Taxi Trip Data (2020-2024)

```python
# Setting the range to get the years/months I want
years = range(2020, 2025)
months = [f"{m:02}" for m in range(1, 13)]

# Base URLs for the two taxi datasets
urls = {
    "yellow": "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_",
    "green": "https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_"
}
```

```python
# Creating a folder for each dataset type, looping through all year/month combinations, and requesting file from server
for color, link in urls.items():
    folder = f"NYC_Taxi_Trip_Data_Parquet/{color}"
    os.makedirs(folder, exist_ok=True)
    
    for y in years:
        for m in months:
            name = f"{color}_tripdata_{y}-{m}.parquet"
            path = f"{folder}/{name}"

            r = requests.get(f"{link}{y}-{m}.parquet")
            if r.ok:
                with open(path, "wb") as f:
                    f.write(r.content)
                print(f"Downloaded: {name}")
            else:
                print(f"Skipped: {name}")
            
            time.sleep(1)
```

### 1.2 Converting to CSV Format using PySpark
_Setting up PySpark locally turned out to be more complex than expected. I spent quite some time testing different versions of Python, Java, and PySpark before realizing I had to adjust the environment variables and set the correct paths. In the end, I got it working with Python 3.10.0, PySpark 4.0.1, JDK 17, and Hadoop 3.3.5 configured as HADOOP_HOME. I am aware that this process could have been easier using tools like Google Colab or pandas for conversion, but I wanted to understand how PySpark works on a system level and make it function properly on my own setup._

```python
# Starting my Spark session
spark = SparkSession.builder.appName("NYC_Parquet_to_CSV").getOrCreate()

# Verifying that my Spark session instance is running
spark
```

```python
# Base folders
base_input_folder = r"NYC_Taxi_Trip_Data_Parquet"
base_output_folder = r"NYC_Taxi_Trip_Data_CSV"

# Creating main output folder
os.makedirs(base_output_folder, exist_ok=True)
```

```python
# Looping through both taxi colors
colors = ["yellow", "green"]
for color in colors:
    input_folder = os.path.join(base_input_folder, color)
    output_folder = os.path.join(base_output_folder, color)
    os.makedirs(output_folder, exist_ok=True)

    print(f"Converting {color}-taxi data...")

    # Looping through all my Parquet files
    for file in os.listdir(input_folder):
        if file.endswith(".parquet"):
            parquet_path = os.path.join(input_folder, file)
            csv_filename = file.replace(".parquet", ".csv")
            csv_path = os.path.join(output_folder, csv_filename)

            # Reading Parquet and write as a single CSV file
            df = spark.read.parquet(parquet_path)
            df.coalesce(1).write.mode("overwrite").option("header", True).csv(csv_path)
            print(f"Converted: {csv_filename}")

print("All files have been successfully converted to CSV")

# Then I will stop the Spark session
spark.stop()
```

#### 2.2.3 Cleaning Numeric Strings/Converting Data Types (grouped by category)

_All columns were initially imported as string in the Yellow dataset, likely due to format issues during download or conversion. Using the TLC 2025 metadata as a reference, I converted each column back to a suitable data type - timestamps for time fields, integers for IDs, strings for flags, and doubles for continuous numeric values. I also removed the column ehail_fee from the Green Taxi dataset because it was mostly NULLs, and not part of the metadata._

```python
# Here I drop ehail_fee, make sure "1.0" -> "1", and cast columns to appropriate datatypes matching the metadata
def cast_column_types(sdf, cfg):
    if "drop_cols" in cfg:
        sdf = sdf.drop(*[c for c in cfg["drop_cols"] if c in sdf.columns])
    for c in cfg["int_cols"]:
        sdf = sdf.withColumn(c, F.regexp_replace(F.trim(F.col(c)), r"\.0$", "").cast(IntegerType()))
    for c in cfg["float_cols"]:
        sdf = sdf.withColumn(c, F.col(c).cast(DoubleType()))
    for c in cfg["timestamp_cols"]:
        sdf = sdf.withColumn(c, F.to_timestamp(F.col(c)))
    sdf = sdf.withColumn("store_and_fwd_flag", F.col("store_and_fwd_flag").cast(StringType()))
    return sdf
```

```python
# In this function I built an expression for all columns so it checks missing/null in one go
def check_missing_values(sdf):
    exprs = []
    for c in sdf.columns:
        dtype = sdf.schema[c].dataType.simpleString()
        if dtype in ["double", "float"]:
            exprs.append(F.sum(F.when(F.isnan(c) | F.col(c).isNull(), 1).otherwise(0)).alias(c))
        else:
            exprs.append(F.sum(F.when(F.col(c).isNull(), 1).otherwise(0)).alias(c))

    sdf.agg(*exprs).show(truncate=False)
    return sdf
```

```python
# Applying the function for Yellow
yellow_step3 = check_missing_values(yellow_step2)
```

```python
# Applying the function for Green
green_step3 = check_missing_values(green_step2)
```

#### 2.2.5 Filtering Data (Temporal, Distance, Fare & Charges)

```python
# I will only keep trips from 2020-2024 with valid drop-off times, realistic distance, fare, and charge values
def filter_temporal_and_numeric(sdf, cfg):
    sdf = (
        sdf
        .filter((F.year(F.col(cfg["pickup_col"])).between(2020, 2024)) &
                (F.year(F.col(cfg["dropoff_col"])).between(2020, 2024)) &
                (F.col(cfg["dropoff_col"]) > F.col(cfg["pickup_col"])))
        .filter((F.col("trip_distance") > 0) & (F.col("trip_distance") <= 100) &
                (F.col("fare_amount") > 0) & (F.col("fare_amount") <= 500) &
                (F.col("total_amount") > 0) & (F.col("total_amount") <= 600))
        .filter((F.col("extra").between(0, 100)) &
                (F.col("mta_tax").between(0, 10)) &
                (F.col("tip_amount").between(0, 200)) &
                (F.col("tolls_amount").between(0, 100)) &
                (F.col("congestion_surcharge").isNull() | (F.col("congestion_surcharge") >= 0)) &
                (F.col("improvement_surcharge") >= 0)))
    return sdf
```

```python
# Applying the function for Yellow with additional filter on airport_fee
yellow_step4 = filter_temporal_and_numeric(yellow_step3, cfg_yellow) \
    .filter(F.col("airport_fee").isNull() | (F.col("airport_fee") >= 0))
```

```python
# Applying the function for Green
green_step4 = filter_temporal_and_numeric(green_step3, cfg_green)
```

#### 2.2.6 Handling Passenger Count, Missing Fields & Categorical Validation

```python
# Ensuring passenger count is within realistic limits (1-6) (null/0 = 1), dropping/filling missing fields and validating categorical codes
def handle_missing_and_categorical(sdf, cfg):
    sdf = (
        sdf
        .withColumn("passenger_count", F.when((F.col("passenger_count").isNull()) | (F.col("passenger_count") == 0), 1).otherwise(F.col("passenger_count")))
        .filter((F.col("passenger_count") >= 1) & (F.col("passenger_count") <= 6))
        .na.drop(subset=cfg["mandatory_cols"])
        .fillna(cfg["fillna_dict"])
        .filter(F.col("VendorID").isin(cfg["valid_vendors"]))
        .filter(F.col("RatecodeID").isin(cfg["valid_ratecodes"]))
        .filter(F.col("payment_type").isin(cfg["valid_payment_types"]))
    )
    return sdf
```

#### 4.2.3 Average Fare Amount by Distance Intervals (Binned)

_I created seven distance bins for trips up to 100 miles (since I removed all longer trips during cleaning). Then I added a bin_order column to ensure the bins sort correctly. I think this makes it easier to see how fare levels increase with trip distance._

```python
# Here I will first define a function to add the distance bins
def add_distance_bin(df):
    return df.withColumn(
        "distance_bin",
        F.when((F.col("trip_distance") >= 0) & (F.col("trip_distance") < 1),  "0-1 miles")
         .when((F.col("trip_distance") >= 1) & (F.col("trip_distance") < 2),  "1-2 miles")
         .when((F.col("trip_distance") >= 2) & (F.col("trip_distance") < 5),  "2-5 miles")
         .when((F.col("trip_distance") >= 5) & (F.col("trip_distance") < 10), "5-10 miles")
         .when((F.col("trip_distance") >= 10) & (F.col("trip_distance") < 20), "10-20 miles")
         .when((F.col("trip_distance") >= 20) & (F.col("trip_distance") < 50), "20-50 miles")
         .otherwise("50-100 miles")
    )
```

```python
# Then I will define a function to add an order column for the distance bins
def add_bin_order(df):
    return df.withColumn(
        "bin_order",
        F.when(F.col("distance_bin") == "0-1 miles", 0)
         .when(F.col("distance_bin") == "1-2 miles", 1)
         .when(F.col("distance_bin") == "2-5 miles", 2)
         .when(F.col("distance_bin") == "5-10 miles", 3)
         .when(F.col("distance_bin") == "10-20 miles", 4)
         .when(F.col("distance_bin") == "20-50 miles", 5)
         .otherwise(6)
    )
```

```python
# Adding both functions for Yellow
yellow_binned = add_distance_bin(yellow_clean_sdf)
yellow_binned = add_bin_order(yellow_binned)

# Calculating average fare per distance bin
yellow_avg_fare = (
    yellow_binned
    .groupBy("distance_bin", "bin_order")
    .agg(F.round(F.avg("fare_amount"), 2).alias("avg_fare"))
    .orderBy("bin_order")
    .drop("bin_order")
)

yellow_avg_fare.show(truncate=False)
```

```python
# Adding both functions for Green
green_binned = add_distance_bin(green_clean_sdf)
green_binned = add_bin_order(green_binned)

# Calculating average fare per distance bin
green_avg_fare = (
    green_binned
    .groupBy("distance_bin", "bin_order")
    .agg(F.round(F.avg("fare_amount"), 2).alias("avg_fare"))
    .orderBy("bin_order")
    .drop("bin_order")
)

green_avg_fare.show(truncate=False)
```

```python
# Function to add time-based features
def add_time_features(df, pickup_col):
    df = (
        df
        # Hour of day (0-23)
        .withColumn("pickup_hour", F.hour(F.col(pickup_col)))

        # Day of week (1=Sunday, 7=Saturday)
        .withColumn("pickup_dayofweek", F.dayofweek(F.col(pickup_col)))

        # Weekend (Sat=7, Sun=1) -> 1 = weekend, 0 = weekday
        .withColumn("is_weekend", F.when(F.col("pickup_dayofweek").isin([1, 7]), 1).otherwise(0))

        # Month (1-12)
        .withColumn("pickup_month", F.month(F.col(pickup_col)))
    )

    return df
```

### 5.2 Calculating Trip Duration

_I created trip_duration_minutes by converting the pickup and dropoff timestamps to UNIX seconds and dividing by 60. I chose this name to keep the feature easy to understand. I also rounded the values to two decimals to make the output more readable during my checks, but the model would work just as well without rounding since Spark ML handles continuous values naturally. The rounding was therefore mainly a presentation choice rather than something that affects the predictive performance. I also used a small safety filter to remove any negative durations just in case, even though these were already handled earlier in the cleaning process._

```python
# Function to calculate trip duration in minutes
def add_trip_duration(df, pickup_col, dropoff_col):
    df = df.withColumn("trip_duration_minutes", F.round((F.unix_timestamp(F.col(dropoff_col)) - F.unix_timestamp(F.col(pickup_col))) / 60.0, 2))
    
    # Keeping only positive durations (Even though my cleaning should have ensured this)
    df = df.filter(F.col("trip_duration_minutes") > 0)

    return df
```

#### 6.2.1 Evaluation Setup

_I used RMSE, MAE, and R^2 together because each of them highlights different aspects of performance:_

- RMSE is measured in my fare units, but because it squares the errors it penalises large mistakes much more strongly than MAE.

- MAE gives a more intuitive “average deviation” in fare units.

- R^2 shows how much variation the model actually explains.

_Using all three felt like a more balanced and fair way of evaluating the models, especially on noisy real-world taxi data._

```python
# Defining evaluators once to reuse
evaluator_rmse = RegressionEvaluator(metricName="rmse", labelCol="fare_amount", predictionCol="prediction")
evaluator_mae = RegressionEvaluator(metricName="mae", labelCol="fare_amount", predictionCol="prediction")
evaluator_r2 = RegressionEvaluator(metricName="r2", labelCol="fare_amount", predictionCol="prediction")
```

```python
# Function to evaluate my predictions
def evaluate_predictions(model_name, predictions):
    rmse = evaluator_rmse.evaluate(predictions)
    mae = evaluator_mae.evaluate(predictions)
    r2 = evaluator_r2.evaluate(predictions)

    return {"Model": model_name, "RMSE": rmse, "MAE": mae, "R2": r2}
```

#### 6.2.2 Training Loop with Timing
_I chose to measure training time as part of the evaluation because I assume that performance is not only about predictive accuracy. With large datasets, I would say that speed and scalability matter just as much, and I wanted to show that I am aware of those trade-offs._

```python
# Defining dictionaries so I can store the trained models/prediction for the visuals later on
yellow_models = {}
green_models = {}
```

```python
# Here I define a function to train and evaluate a model, measuring training time
def train_and_evaluate(model, train_df, test_df, model_name):
    start = time.time()
    fitted = model.fit(train_df)

    # Storing only the 2 needed columns to save memory
    predictions = fitted.transform(test_df).select("fare_amount", "prediction")
    
    elapsed = time.time() - start
    metrics = evaluate_predictions(model_name, predictions)
    metrics["Training Time (min)"] = round(elapsed / 60, 2)

    # Returning the fitted models + predictions + metrics
    return fitted, predictions, metrics
```

### 6.3 Plotting Actual vs Predicted Fare Amounts

_To avoid overplotting and unnecessary memory usage, I needed to sample 5,000 prediction rows for the scatter plots. Since the model evaluation (RMSE/MAE/R^2) is performed on the full test dataset, sampling should not affect any results but just improve visual clarity and make my plots more readable._

```python
# Setting seaborn theme for consistency
sns.set_theme(style="whitegrid", font_scale=1.15)
```

```python
# Converting a Spark DF sample to pandas
def to_pandas(pred_df, sample_size=5000):

    total = pred_df.count()
    fraction = min(1.0, sample_size / total)

    sample = pred_df.sample(withReplacement=False, fraction=fraction, seed=42)

    pdf = sample.toPandas()
    return pdf
```

```python
# Plotting function for Actual vs Predicted
def plot_actual_vs_pred(pdf, model_name, dataset_name):
    plt.figure(figsize=(7, 6))
    sns.scatterplot(data=pdf, x="fare_amount", y="prediction", alpha=0.35, s=22, color="royalblue")
    
    # Adding perfect fit line and regression line
    max_val = max(pdf["fare_amount"].max(), pdf["prediction"].max())
    
    plt.plot([0, max_val], [0, max_val], "--", color="darkred", linewidth=2, label="Perfect Fit")
    sns.regplot(data=pdf, x="fare_amount", y="prediction", scatter=False, color="black", line_kws={"linewidth": 1.4, "alpha": 0.7})
    plt.title(f"{dataset_name}: {model_name}", fontsize=15, weight="bold")
    plt.xlabel("Actual Fare Amount")
    plt.ylabel("Predicted Fare Amount")
    plt.tight_layout()
    plt.show()
```

```python
# Plotting for Yellow - Showing only Linear Regression (see markdown texts for explanation)
for model_name, (fitted, preds) in yellow_models.items():
    pdf = to_pandas(preds)
    plot_actual_vs_pred(pdf, model_name, "Yellow Taxi")
```
